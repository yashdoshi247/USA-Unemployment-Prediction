```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

sns.set()
```

```python
raw_data = pd.read_csv("C:/Users/yashd/Downloads/archive (1)/unemployment_data_us.csv")
df = raw_data.copy()
df.head(10)
```

```python
df.shape
```

```python
df_2019 = df[df['Year'] <= 2019]
df_2019.info()
```

```python
df_2019['Date'] =pd.to_datetime(df_2019.Date)
df_2019 = df_2019.sort_values(by='Date', ascending=True)
```

```python
plt.figure(figsize=(16,6))
plt.plot(df_2019['Date'], df_2019['Primary_School'])
plt.plot(df_2019['Date'], df_2019['High_School'])
plt.plot(df_2019['Date'], df_2019['Associates_Degree'])
plt.plot(df_2019['Date'], df_2019['Professional_Degree'])
plt.title("Unemployment rate by degree", fontsize = 20)
plt.legend(["Primary School","High School","Associate Degree","Professional Degree"], loc = "upper right")
plt.xlabel("Date", fontsize = 14)
plt.ylabel("Unemployment rate", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(16, 6))
plt.plot(df_2019['Date'], df_2019['Asian'])
plt.plot(df_2019['Date'], df_2019['Hispanic'])
plt.plot(df_2019['Date'], df_2019['White'])
plt.plot(df_2019['Date'], df_2019['Black'])
plt.title("Unemployment rate by race", fontsize=20)
plt.legend(["Asian", "Hispanic", "White", "Black"], loc="upper right")
plt.xlabel("Date", fontsize=14)
plt.ylabel("Unemployment rate", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(16, 6))
plt.plot(df_2019['Date'], df_2019['Men'])
plt.plot(df_2019['Date'], df_2019['Women'])
plt.title("Unemployment rate by gender", fontsize=20)
plt.legend(["Men", "Women"], loc="upper right")
plt.xlabel("Date", fontsize=14)
plt.ylabel("Unemployment rate", fontsize=14)
plt.show()
```
